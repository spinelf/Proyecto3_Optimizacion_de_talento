## Decidimos quitar columnas que no nos sirven

```python
# Eliminar las columnas 'id' y 'same_as_monthly_income'
data_noid = data.drop(columns=["same_as_monthly_income","employee_count","salary","standard_hours", "over_18",'performance_rating',"number_children","years_in_current_role"])

# Mostrar el DataFrame resultante
data_noid
```

### Lo que hay que cambiar de la 0 a la 21

- daily_rate: quitar símbolo $ y convertir a int

```python
# Aseguramos que los valores de 'daily_rate' son de tipo string
data_noid['daily_rate'] = data_noid['daily_rate'].astype(str)

# Lista de símbolos a eliminar en los valores de la columna
quitar_simbolos = [' ', '$']

# Itera sobre cada símbolo en la lista de símbolos a eliminar
for simbolo in quitar_simbolos:
    # Elimina el símbolo actual de todos los valores de la columna 'daily_rate'
    data_noid['daily_rate'] = data_noid['daily_rate'].str.replace(simbolo, '', regex=False)

# Rellena los valores NaN y los valores vacíos con '0'
data_noid['daily_rate'] = data_noid['daily_rate'].replace(['nan', ''], '0')

# Reemplazamos las comas por puntos
data_noid['daily_rate'] = data_noid['daily_rate'].str.replace(',', '.')

# Convertimos la columna a tipo numérico (dtype: "object" pasa a ser dtype: "float64")
data_noid['daily_rate'] = pd.to_numeric(data_noid['daily_rate'], errors='coerce') 

# Cambiamos los valores de tipo string a int usando ".astype(int)"
data_noid['daily_rate'] = data_noid['daily_rate'].astype("int32")
```

- `age` es `object` y debería ser `int`
- `hourly_rate` aparece como `object` y debería ser `int`
- `monthly_income`: convertir a `int`

```python
columnas3 = ['age', 'hourly_rate']

# Itera sobre cada columna en la lista de columnas a convertir
for columna in columnas3:
    # Asegúrate de que los valores de la columna son de tipo string
    data_noid[columna] = data_noid[columna].astype(str)
    # Elimina cualquier valor no numérico o vacío y reemplaza por 0
    data_noid[columna] = data_noid[columna].replace(['nan', '', ' '], '0')
    # Convierte la columna a tipo numérico (int)
    data_noid[columna] = pd.to_numeric(data_noid[columna], errors='coerce').fillna(0).astype("int32")

# Muestra las primeras filas del DataFrame para verificar que los valores de las columnas se han modificado correctamente
data_noid.head(2)
```

- cambiar todos los `valores` a `minúsculas`

```python
# Lista de columnas a convertir a minúsculas de al 0 a la 21
columns_to_lower = ['age', 'attrition', 'business_travel', 'daily_rate', 'department',
       'distance_from_home', 'education', 'education_field', 'employee_number',
       'environment_satisfaction', 'gender', 'hourly_rate', 'job_involvement',
       'job_level', 'job_role', 'job_satisfaction', 'marital_status',
       'monthly_income', 'monthly_rate', 'num_companies_worked',
       'overtime', 'percent_salary_hike',
       'relationship_satisfaction', 'stock_option_level',
         'training_times_last_year', 'work_life_balance',
       'years_at_company', 
       'years_since_last_promotion', 'years_with_curr_manager', 'year_birth',
       'role_department', 'remote_work']

# Convertir a minúsculas las columnas especificadas
for column in columns_to_lower:
    if data_noid[column].dtype == 'object':  # Verifica si la columna es de tipo str
        data_noid[column] = data_noid[column].str.lower()
        
data_noid.head(5)
```

- quitar los `NaN` de las columnas que lo tengan

```python
# Reemplazar NaN en las columnas especificadas por 'N/A'
columns_to_replace = ['business_travel', 'department', 'education_field', 'marital_status','overtime','work_life_balance']
replacement_value = 'N/A'

data_noid[columns_to_replace] = data_noid[columns_to_replace].fillna(replacement_value)

# Mostrar los primeros registros para verificar
data_noid.head(2)
```

```python
#Quitar los valores negativos y reemplazarlos por "N/A" y convertir la columna a INT
data_noid["distance_from_home"] = data_noid["distance_from_home"].apply(lambda x: "N/A" if x < 0 else x)

data_noid["distance_from_home"] = pd.to_numeric(data_noid["distance_from_home"], errors='coerce').fillna(0).astype("int32")
```

```python
#Reemplazamos los 0 por male y 1 por female. Hemos tomado decisión de usar 0 para male y 1 para female siguiendo la logica explicada en la fase 2 del gitbook.
#Primero cambiamos la columna a tipo object para poder añadir el sexo.
data_noid["gender"] = data_noid["gender"].astype(str)

#male
data_noid["gender"] = data_noid["gender"].replace(['0'], 'male')

#female
data_noid["gender"] = data_noid["gender"].replace(['1'], 'female')

#Verificamos los datos unicos
data_noid["gender"].unique()
```

```python
#Reemplazamos el marreid por married dentro de la columna
data_noid["marital_status"] = data_noid["marital_status"].replace(['marreid'], 'married')

#Verificamos que esten los datos unicos
data_noid["marital_status"].unique()
```

```python
data_noid["monthly_income"] = data_noid["monthly_income"].fillna('0') #Reemplazar los nan por un 0 para poder luego hacer el rango

data_noid["monthly_income"] = data_noid["monthly_income"].str.replace(',', '.') #conversión de los datos cambiando las comas por punto

data_noid["monthly_income"] = data_noid["monthly_income"].astype(float) # Convertir la columna "monthly income" a tipo float
```

```python
# Convertir la columna 'employee_number' a numérico, forzando errores a NaN, y luego a entero
data['employee_number'] = pd.to_numeric(data['employee_number'].str.replace(',', ''), errors='coerce').fillna(0).astype(int)

# Primero tengo que filtrar por los números que ya existen
numeros_actuales = data[data["employee_number"] != 0]["employee_number"]

# Busco el valor máximo como principio de mi nuevo listado de números
max_numeros_actuales = numeros_actuales.max()

# Vamos a buscar cuántos empleados tienen un cero 
num_zeros = data[data["employee_number"] == 0].shape[0]

# Lista de nuevos números
nuevos_numeros = list(range(max_numeros_actuales + 1, max_numeros_actuales + 1 + num_zeros))

# Añado los números para todos los empleados que tengan un cero
data.loc[data["employee_number"] == 0, "employee_number"] = nuevos_numeros

print("DataFrame actualizado:")
data

```
